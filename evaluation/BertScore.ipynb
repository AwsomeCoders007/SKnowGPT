```python
from evaluate import load
import csv
import os
dir_path = os.getcwd()

input_file = '../output/SKnowGPT_output_without_dtt_cot1_pruned_DisTreatKG_threshold_gpt3.csv'
output_file = f'{dir_path}/BERTScore_results/SKnowGPT_output_without_dtt_cot1_pruned_DisTreatKG_threshold_gpt3_bertscore.csv'

bertscore = load("bertscore")

with open(input_file,'r',newline="") as f_input , open(output_file, 'a+', newline='') as f_output:
  reader = csv.reader(f_input)
  writer = csv.writer(f_output)

  header = next(reader)
  header.extend(["output1_precision","output1_recall","output1_f1"])
  writer.writerow(header)

  for i,row in enumerate(reader):
    print('--',i+1,'--')

    output1_text = [row[3].strip("\n")]
    
    references = [row[2].strip()]
    references = [references[0].strip('[]').strip('\'')]
    
    results1 = bertscore.compute(predictions=output1_text, references=references, model_type="distilbert-base-uncased")
      
    print(results1["precision"][0],results1["recall"][0],results1["f1"][0])
    
    row.extend([results1["precision"][0],results1["recall"][0],results1["f1"][0]])
    writer.writerow(row)
    
```
